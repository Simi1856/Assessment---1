**1.** Create a DataFrame `df` from this dictionary `data` which has the index `labels`.

````python
import pandas as pd
import numpy as np

# Data dictionary
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}

# Labels list
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

# Create DataFrame
df = pd.DataFrame(data, index=labels)

# Display the DataFrame
df
````

**18.** Load the ny-flights dataset to Python

````python
import numpy as np
import seaborn as sns
import pandas as pd

flights = pd.read_csv('/content/ny-flights.csv')

flights.head()
````

**19.** Which airline ID is present maximum times in the dataset

````python
airline_counts = flights['unique_carrier'].value_counts()
max_airline_id = airline_counts.idxmax()
max_count = airline_counts.max()
print(f"The airline ID present maximum times is '{max_airline_id}' with {max_count} occurrences.")
````

**20.** Draw a plot between dep_delay and arr_delay

````python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.scatterplot(data=flights, x='dep_delay', y='arr_delay')
plt.title('Scatter Plot of Departure Delay vs Arrival Delay')
plt.xlabel('Departure Delay (minutes)')
plt.ylabel('Arrival Delay (minutes)')
plt.show()
````
